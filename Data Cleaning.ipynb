# Data Cleaning

### Imports

```python
import numpy as np
import pandas as pd 
import warnings
warnings.filterwarnings('ignore')
import math
import Settings
client_id = Settings.CLIENT_ID
secret = Settings.CLIENT_PASS
import requests
```

### Geting API access_token

```python
headers = {'Content-Type': 'application/x-www-form-urlencoded'}
payload = {'grant_type' : 'client_credentials', 'client_id': client_id, 'client_secret': secret}
r = requests.post("https://accounts.spotify.com/api/token", data=payload, headers=headers)
AT , ATtype = r.json()['access_token'] , r.json()['token_type']
```

### Loading Data

```python
df_db = pd.read_csv('top_10000_1960-now.csv')
df_db
```

```python
df_db.info()
```

### Getting just raw ids

```python
df_db['Track URI'] = df_db['Track URI'].map(lambda x: x[14:])
df_db.drop(['Track Preview URL','Album Genres'], axis=1, inplace=True)
df_db.dropna(inplace=True)
```

```python
df_db.info()
```

### Load user's liked playlist

```python
df_user_b = pd.read_csv('Ever Liked.csv')
df_user_b
```

```python
df_user_b.info()
```

### More Loading Data

```python
df_giga = pd.read_csv('tracks_features.csv')
df_giga.info()
```

### Fixing the artists column

```python
df_giga['artists'] = df_giga['artists'].map(lambda x: x.strip('[]').split(',')[0] if len(x.strip('[]').split(',')) == 1 else x.strip('[]').split(','))
df_giga['artists'] = df_giga['artists'].map(lambda x: [i.strip("'") for i in x]  if type(x) is list else x.strip("'"))
df_giga['artists']
```

###  Combinding data 

```python
df = pd.DataFrame()
```

```python
df['id'] = list(df_db['Track URI']) + list(df_giga['id'])
df['Title'] = list(df_db['Track Name']) +  list(df_giga['name']) 
df['Artist(s)'] = list(df_db['Artist Name(s)']) + list(df_giga['artists'])
df['Acousticness'] = list(df_db['Acousticness'])  + list(df_giga['acousticness'])
df['Danceability'] = list(df_db['Danceability'])  + list(df_giga['acousticness'])
df['Duration (ms)'] = list(df_db['Track Duration (ms)']) + list(df_giga['duration_ms'])
df['Energy'] = list(df_db['Energy']) + list(df_giga['energy'])
df['Instrumentalness'] = list(df_db['Instrumentalness']) + list(df_giga['instrumentalness'])
df['Key'] = list(df_db['Key'])  + list(df_giga['key'])
df['Liveness'] = list(df_db['Liveness']) + list(df_giga['liveness'])
df['Loudness'] = list(df_db['Loudness']) + list(df_giga['loudness'])
df['Mode'] = list(df_db['Mode']) + list(df_giga['mode'])
df['Speechiness'] =  list(df_db['Speechiness']) + list(df_giga['speechiness'])
df['BPM'] = list(df_db['Tempo']) + list(df_giga['tempo'])
df['Time_Signature'] = list(df_db['Time Signature']) + list(df_giga['time_signature'])
df['Valence'] = list(df_db['Valence']) + list(df_giga['valence'])
df
```

### Dropping nulls

```python
df.dropna(inplace=True)
```

### Getting the meterics for the user's liked playlist using the API

```python
df_2B_append = pd.DataFrame()
df_2B_append['id'] = df_user_b['Spotify Track Id']
```

```python
headers = {'Authorization': ATtype + " " + AT}
arts, title, pop = [] , [] , []
lst = list(df_user_b['Spotify Track Id'])
for id in [lst[i:i + 50] for i in range(0, len(lst), 50)]:
    ids = '%2C'.join(id)
    r = requests.get("https://api.spotify.com/v1/tracks?ids="+ids, headers=headers)
    for t in r.json()['tracks']:
        if t is None:
            art = np.nan
            title.append(np.nan)
            pop.append(np.nan)
        else:
            art = t['artists']
            if len(art) == 1: arts.append(art[0]['name'])
            else: arts.append([x['name'] for x in art])
            title.append(t['name'])
            pop.append(t['popularity'])
df_2B_append['Title'], df_2B_append['Artist(s)'], df_2B_append['Popularity'] = title, arts, pop
df_2B_append[['Title', 'Artist(s)','Popularity']]
```

```python
headers = {'Content-Type': 'application/x-www-form-urlencoded'}
payload = {'grant_type' : 'client_credentials', 'client_id': client_id, 'client_secret': secret}
r = requests.post("https://accounts.spotify.com/api/token", data=payload, headers=headers)
AT , ATtype = r.json()['access_token'] , r.json()['token_type']
headers = {'Authorization': ATtype + " " + AT}
```

```python
acousticness, danceability, duration_ms, energy, instrumentalness, key, liveness, loudness, mode, speechiness, tempo, time_signature, valence  \
= [] , [] , [], [] , [] , [], [] , [] , [], [] , [] , [], []
for id in [lst[i:i + 50] for i in range(0, len(lst), 50)]:
    ids = '%2C'.join(id)
    r = requests.get("https://api.spotify.com/v1/audio-features?ids="+ids, headers=headers)
    for t in r.json()['audio_features']:
        if t is None:
            acousticness.append(np.nan)
            danceability.append(np.nan)
            duration_ms.append(np.nan)
            energy.append(np.nan)
            instrumentalness.append(np.nan)
            key.append(np.nan)
            liveness.append(np.nan)
            loudness.append(np.nan)
            mode.append(np.nan)
            speechiness.append(np.nan)
            tempo.append(np.nan)
            time_signature.append(np.nan)
            valence.append(np.nan)
        else:
            acousticness.append(t['acousticness'])
            danceability.append(t['danceability'])
            duration_ms.append(t['duration_ms'])
            energy.append(t['energy'])
            instrumentalness.append(t['instrumentalness'])
            key.append(t['key'])
            liveness.append(t['liveness'])
            loudness.append(t['loudness'])
            mode.append(t['mode'])
            speechiness.append(t['speechiness'])
            tempo.append(t['tempo'])
            time_signature.append(t['time_signature'])
            valence.append(t['valence'])
df_2B_append['Acousticness'], df_2B_append['Danceability'], df_2B_append['Duration (ms)'], df_2B_append['Energy'],\
df_2B_append['Instrumentalness'], df_2B_append['Keys'], df_2B_append['Liveness'], df_2B_append['Loudness'], df_2B_append['Mode'],\
df_2B_append['Speechiness'], df_2B_append['BPM'], df_2B_append['Time_Signature'], df_2B_append['Valence'] =  acousticness, \
danceability, duration_ms, energy, instrumentalness, key, liveness, loudness, mode, speechiness, tempo, time_signature, valence
```

### Using the data from api we can put the user's data into the main data frame

```python
df = pd.concat([df, df_2B_append], ignore_index=True)
df
```

### Fixing the artists column

```python
temp = df
temp['Artist(s)'] = temp['Artist(s)'].map(lambda x: str(x))
temp.drop_duplicates(subset=['Artist(s)', 'Title'], inplace=True)
temp['Artist(s)'] = temp['Artist(s)'].map(lambda x: x.split(',')[0] if len(x.split(',')) == 1  else x.split(','))
df = temp
df 
```

```python
df.dropna(inplace=True)
```
